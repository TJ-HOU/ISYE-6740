```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy.sparse.linalg as ll
```

```python
"""It helps visualising the portraits from the dataset."""
def plot_portraits(images, titles, h, w, n_row, n_col):
    plt.figure(figsize=(2.2 * n_col, 2.2 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.20)
    
    total = images.shape[0]
    for i in range(total):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i])
        plt.xticks(())
        plt.yticks(())
```

```python
dir='data/yalefaces'
photos=os.listdir(dir)[0:1001]
photos = sorted(photos)
images=[dir+'/' + photo for photo in photos]
images=np.array([plt.imread(image) for image in images], dtype=np.float64)
names=[name for name in photos]
n_samples, h, w = images.shape
plot_portraits(images, names, h, w, n_row=4, n_col=6)
```

```python
images[0].shape
```

```python
rows = np.array([i for i in range(0, 243, 4)])
columns = np.array([i for i in range(0, 320, 4)])
images_scaled = np.empty((22, 61, 80), dtype=np.float64)

i = 0
for image in images:
    temp = images[i][rows, :]
    temp = temp[:, columns]
    images_scaled[i] = temp
    i = i+1

#images_scaled
n_samples, h, w = images_scaled.shape
plot_portraits(images_scaled, names, h, w, n_row=4, n_col=6)
 
print(images_scaled.shape)
np.reshape(images_scaled[0], -1).shape
```

```python
X = np.empty((22, 4880), dtype=np.float64)
i = 0
for image in images_scaled :
    X[i] = np.reshape(images_scaled[i], -1)
    i = i+1
    
print(images_scaled.shape)
print(X.shape)
print(X)
print(X.T)
```

```python
X_01 = np.delete(X, [i for i in range(11, 22)], axis=0)
X_01 = np.delete(X_01, [0], axis=0)
names_01 = np.delete(names, [i for i in range(11, 22)], axis=0)
names_01 = np.delete(names_01, [0], axis=0)

X_01_test = np.delete(X, [i for i in range(1, 22)], axis=0)
names_01_test = np.delete(names, [i for i in range(1, 22)], axis=0)

X_14 = np.delete(X, [i for i in range(0, 11)], axis=0)
X_14 = np.delete(X_14, [0], axis=0)
names_14 = np.delete(names, [i for i in range(0, 11)], axis=0)
names_14 = np.delete(names_14, [0], axis=0)

X_14_test = np.delete(X, [i for i in range(0, 11)], axis=0)
X_14_test = np.delete(X_14_test, [i for i in range(1, 11)], axis=0)
names_14_test = np.delete(names, [i for i in range(0, 11)], axis=0)
names_14_test = np.delete(names_14_test, [i for i in range(1, 11)], axis=0)

names_14
print(X_14_test.shape)
```

```python
tempX_14 = np.dot(X_14.T, X_14)
K = 6
S,W = ll.eigs(tempX_14,k = K)

temp = np.empty((K, 61, 80), dtype=np.float64)

vnames = []
i = 0
for v in W.T: 
    temp[i] = np.reshape(v, (61, 80))
    i = i+1
    vnames.append("vector" +str(i))
    
n_samples, h, w = temp.shape
print(temp.shape)
plot_portraits(temp, vnames, h, w, n_row=2, n_col=6)

```

```python
tempX_01 = np.dot(X_01.T, X_01)
K = 6
S,W = ll.eigs(tempX_01,k = K)

temp = np.empty((K, 61, 80), dtype=np.float64)

vnames = []
i = 0
for v in W.T: 
    temp[i] = np.reshape(v, (61, 80))
    i = i+1
    vnames.append("vector" +str(i))
    
n_samples, h, w = temp.shape
print(temp.shape)
plot_portraits(temp, vnames, h, w, n_row=2, n_col=6)

```
