```python
import numpy as np
import numpy.matlib
from skimage import io
from matplotlib import pyplot as plt
import random
import scipy.misc
```

```python
image = io.imread('HW1_Data/beach.bmp')
io.imshow(image)
io.show()
```

```python
rows = image.shape[0]
cols = image.shape[1]
image = image/255
X = image.reshape(image.shape[0]*image.shape[1],3)

K = 16 #number of clusters
max_iters = 50 #number of times the k-mean should run
```

```python
def init_medoids(X,K):
    c = random.sample(list(X),K)
    return c
```

```python
def closest_medoids(X,c):
    K = np.size(c,0)
    idx = np.zeros((np.size(X,0),1))
    arr = np.empty((np.size(X,0),1))

    for i in range(0,K):
        y = c[i]
        temp = np.ones((np.size(X,0),1))*y
        b = np.power(np.subtract(X,temp),2)
        a = np.sum(b,axis = 1) #The np.sum gives horizontal array. We need to resize it.
        #a = np.asarray(a)
        a.resize((np.size(X,0),1))
        #print(np.shape(a))
        arr = np.append(arr, a, axis=1)
        #print('shape arr for i={} '.format(i), arr.shape)

    arr = np.delete(arr,0,axis=1)
    idx = np.argmin(arr, axis=1) #The np.argmin give horizontal array
    #print('shape of idx at the end ' , np.shape(idx))
    return idx
```

```python
def update_medoids(X,idx,K):
    n = np.size(X,1) #in our case is always 3 (RGB)
    medoids = np.zeros((K,n))
    for i in range(0,K):
        
        ci = idx==i
        cluster = X[ci]
        minMediod = np.zeros((1, n))
        minDissimilarity = float("inf")
        for data in cluster:
            candidateMedoid = data
            temp = np.ones((np.size(cluster,0),1))*data
            b = np.power(np.subtract(cluster,temp),2)
            a = np.sum(b,axis = 1)
            if a < minDissimilarity :
                minMediod = data
                minDissimilarity = a
            
        medoids[i] = minMediod
        
    return medoids
```

```python
def run_kMean(X,initial_medoids,max_iters):
    m = np.size(X,0)
    n = np.size(X,1)
    K = np.size(initial_medoids,0)
    medoids = initial_medoids
    previous_medoids = medoids
    idx = np.zeros((m,1))
    for i in range(1,max_iters):
        idx = closest_medoids(X,medoids)
        medoids = update_medoids(X,medoids,idx,K)
    return medoids,idx
```
